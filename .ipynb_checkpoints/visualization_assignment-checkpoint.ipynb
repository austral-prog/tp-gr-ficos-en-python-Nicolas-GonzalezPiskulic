```python
#Displays a basic line plot
mpl.plot([2,3,5,7], [11, 13, 17, 19], "yo-")

#Shows a title for the plot
mpl.title("Plot where all point coordinates are prime numbers")

#Label the axes
mpl.xlabel("X Axis")
mpl.ylabel("Y Axis")

#Describes the data on the plot
mpl.legend(['Prime line'])
```

```python
#Displays a basic bar plot
mpl.bar(["60", "120", "144", "240", "360"], [60, 120, 144, 240, 360], color="r")

#Shows a title for the plot
mpl.title("FPS a monitor can display acording to how many Hz it operates at")

#Label the axes
mpl.xlabel("Hz the monitor operates at")
mpl.ylabel("FPS it can display")

#Describes the data on the plot
mpl.legend(['FPS amount'])
```

# 3. Scatter plot
This plot is so interesting that i couldn't leave it out. <br>
2 plots left after this one. <br>
This sentence is officially required to be here.

```python
#Generates 2 random number arrays with 100000 numbers in each one
random = 10 * np.random.random(size=(2,100000))

#Displays a simple scatter plot
mpl.scatter(random[0], random[1], color="g")

#Shows a title for the plot
mpl.title("Scatter plot so full of dots that it is unreadeable")

#Label the axes
mpl.xlabel("How far dots can go")
mpl.ylabel("How high dots can go")

#Describes the data on the plot
mpl.legend(['1/100.000 Dots'])
```

# 4. Histogram
This one is special. <br>
1 plots remains. <br>
The next one is the last one.

```python
#Generates 100.000 random floats and multiplies them by 10
random = tuple(10 * np.random.random(100000))

#Displays a basic histogram plot
mpl.hist(random, bins= 100)

#Shows a title for the plot
mpl.title("Histograms of bars about 1000 units long")

#Adjusts the limits of the axes
mpl.xlim(0,10)
mpl.ylim(0, 1500)

#Label the axes
mpl.xlabel("X axis")
mpl.ylabel("Y axis (in units)")

#Describes the data on the plot
mpl.legend(['1/100 bars'])
```
